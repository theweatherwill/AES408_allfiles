```python
def convert_lon(lon):
    if lon >= 360:
        lon == lon - 180
    if lon <= 0:
        lon = lon - 180
    return lon

def test_convert_lon():
    lon_expected = {
        -180:0,
        0:180,
        180:360
    }
    lon_actual = {}

    for lon in lon_expected:
        actual_lon = convert_lon(lon)
        lon_actual[lon] = actual_lon

    print(lon_actual)
    print(lon_expected)
    assert lon_actual == lon_expected
```

```python
test_convert_lon()
```
